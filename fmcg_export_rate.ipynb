The question: Which destination or which country received the most product?


```python
import pandas as pd
```

```python
df = pd.read_csv("./ExportSales.csv")
df.head()
```

```python
df["Destenation"].value_counts()
```

```python
print(df["Destenation"].value_counts()["IRQ"])
```

```python
sudlortonggarncountry = df.groupby("Destenation")
print(sudlortonggarncountry)
```

```python
df["Destenation"].value_counts()["IRQ"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Destenation', order=df['Destenation'].value_counts().index)
plt.title('Count of Products by Destination')
plt.xlabel('Destination')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
