```python
import pandas as pd
```

```python
df=pd.read_csv("./Chocolate Sales (2).csv")
df.head()
```

```python
df["Product"].value_counts()
```

```python
df.groupby('Product')['Boxes Shipped']
```

```python
df.sort_values(by='Boxes Shipped', ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
country_import = (
    df.groupby("Country")["Boxes Shipped"]
    .sum()
    .sort_values(ascending=False)
)

country_import.head(1)
print(country_import)
```

```python
plt.figure()
plt.bar(country_import.index, country_import.values)
plt.xlabel("Product")
plt.ylabel("Boxes Shipped")
plt.title("Total Chocolate Boxes Shipped by Product")
plt.show()
```
